```python
import time
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import re


#Creating Class

class Jobs:
    jobTitle    = ""
    jobSource   = ""
    jobType     = ""
    jobDesc     = ""

    def __init__(self, jobTitle, jobSource, jobType, jobDesc):
        self.jobTitle = jobTitle
        self.jobSource = jobSource
        self.jobType = jobType
        self.jobDesc = jobDesc

    def showDetail(self):
        print("Title:" + str(self.jobTitle))
        print(  self.jobSource)
        print(  self.jobType)
        print(  self.jobDesc)
        print("***")

#Web Scraping

URL = "https://www.workbc.ca/Jobs-Careers/Find-Jobs/Jobs.aspx"

browser = webdriver.Chrome()
browser.get(URL)

time.sleep(3)

search = browser.find_element_by_css_selector("#p_lt_ctl00_pageplaceholder_p_lt_ctl00_editabletext_ucEditableText_widget1_ctl00_txtKeywordSearch")
SEARCH_TERM = "data+analyst"

search.send_keys(SEARCH_TERM)

# Find the search button - this is only enabled when a search query is entered
button = browser.find_element_by_css_selector("#p_lt_ctl00_pageplaceholder_p_lt_ctl00_editabletext_ucEditableText_widget1_ctl00_btnFindJobsSearch")
button.click()

time.sleep(3)

titleList = []
jobSourceList = []
positionTypeList = []
descriptionList = []
jobList = []
jobListFinal = []

page = 1

for i in range(0, 3):
    titles = browser.find_elements_by_css_selector(".job-title")
    sources = browser.find_elements_by_css_selector(".job-source")
    positionTypes = browser.find_elements_by_css_selector(".job-position-type")
    descriptions = browser.find_elements_by_css_selector(".job-location")

    for title, source, type, desc in zip(titles, sources, positionTypes, descriptions):
        textContent = title.get_attribute('innerHTML')
        soup = BeautifulSoup(textContent, features="lxml")
        jobTitle = soup.get_text().strip()
        jobTitle = re.sub(r"[\n\t]*", "", jobTitle)
        jobTitle = re.sub('[ ]{2,}', '*', jobTitle)
        jobList.append(jobTitle)
        titleList.append(jobTitle)

        textContent = source.get_attribute('innerHTML')
        soup = BeautifulSoup(textContent, features="lxml")
        jobSource = soup.get_text().strip()
        jobSource = re.sub(r"[\n\t]*", "", jobSource)
        jobSource = re.sub('[ ]{2,}', '', jobSource)
        jobList.append(jobSource)
        jobSourceList.append(jobSource)

        textContent = type.get_attribute('innerHTML')
        soup = BeautifulSoup(textContent, features="lxml")
        jobType = soup.get_text().strip()
        jobType = re.sub(r"[\n\t]*", "", jobType)
        jobType = re.sub('[ ]{2,}', '', jobType)
        jobList.append(jobType)
        positionTypeList.append(jobType)

        textContent = desc.get_attribute('innerHTML')
        soup = BeautifulSoup(textContent, features="lxml")
        jobDesc = soup.get_text().strip()
        jobDesc = re.sub(r"[\n\t]*", "", jobDesc)
        jobDesc = re.sub('[ ]{2,}', '', jobDesc)
        jobList.append(jobDesc)
        descriptionList.append(jobDesc)

        job = Jobs (jobTitle, jobSource, jobType, jobDesc)
        jobListFinal.append(job)

        dataSet = zip(titleList, jobSourceList, positionTypeList, descriptionList)

    page += 1
    URL_NEXT = "https://www.workbc.ca/jobs-careers/find-jobs/jobs.aspx?keywords="\
                       + SEARCH_TERM + "&page="
    URL_NEXT = URL_NEXT + str(page)
    browser.get(URL_NEXT)
    

#    print("Count: ", str(i))
    time.sleep(3)

    for job in jobListFinal:
        job.showDetail()


##### Data Frame #####


df = pd.DataFrame(dataSet, columns=['Job Title', 'Source', 'Position Type', 'Description'])
df.to_csv(r'/Users/JUN/Desktop/College/COMP 2454/Data_Analyst.csv', index = False)

path = "/Users/JUN/Desktop/College/COMP 2454/"
csv = "Data_Analyst.csv"
df = pd.read_csv(path + csv, skiprows=1,  encoding = "ISO-8859-1", sep=',',
                   names=('Job Title', 'Source', 'Position Type', 'Description'))

pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)

print(df.head(2))
print(df.tail(2))
```
